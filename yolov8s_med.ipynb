```python
from ultralytics import YOLO

# Load model Ä‘Ã£ huáº¥n luyá»‡n
model = YOLO("/content/drive/MyDrive/runs/y8s_med/y8s_med2/weights/best.pt")  # hoáº·c yolov8n.pt, yolov8s.pt..

results = model("/content/drive/MyDrive/ML/dataset_balanced/train_data/images/test", save=True, conf=0.25)

import os
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO

# ÄÆ°á»ng dáº«n thÆ° má»¥c chá»©a káº¿t quáº£ detect
output_folder = "/content/runs/detect/predict"

# Hiá»ƒn thá»‹ 4 áº£nh Ä‘áº§u tiÃªn
image_files = sorted(os.listdir(output_folder))[20:24]

plt.figure(figsize=(16, 8))
for i, img_name in enumerate(image_files):
    img_path = os.path.join(output_folder, img_name)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(2, 2, i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(img_name)

plt.tight_layout()
plt.show()
```

```python
# prompt: hiá»ƒn thá»‹ Ä‘á»“ thá»‹, hÃ¬nh áº£nh káº¿t quáº£ Ä‘Ã¡nh giÃ¡

# In ra cÃ¡c metric Ä‘Ã¡nh giÃ¡
print(metrics)

# Hiá»ƒn thá»‹ Ä‘á»“ thá»‹ káº¿t quáº£ huáº¥n luyá»‡n
# CÃ¡c file Ä‘á»“ thá»‹ Ä‘Æ°á»£c lÆ°u trong thÆ° má»¥c runs/y8s_med/y8s_medX (X lÃ  sá»‘ láº§n cháº¡y)
# VÃ­ dá»¥: F1_curve.png, confusion_matrix.png, results.png, ...
# TÃ¬m thÆ° má»¥c cháº¡y má»›i nháº¥t
runs_dir = "/content/drive/MyDrive/runs/y8s_med"
run_folders = sorted([f for f in os.listdir(runs_dir) if os.path.isdir(os.path.join(runs_dir, f))])
latest_run_folder = os.path.join(runs_dir, run_folders[-1])
print(f"Latest run folder: {latest_run_folder}")

# Danh sÃ¡ch cÃ¡c file Ä‘á»“ thá»‹ báº¡n muá»‘n hiá»ƒn thá»‹
plot_files = ["results.png", "F1_curve.png", "confusion_matrix.png", "P_curve.png", "R_curve.png"]

plt.figure(figsize=(18, 12))
for i, plot_file in enumerate(plot_files):
    plot_path = os.path.join(latest_run_folder, plot_file)
    if os.path.exists(plot_path):
        img = cv2.imread(plot_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(2, 3, i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(plot_file)
    else:
        print(f"Plot file not found: {plot_path}")

plt.tight_layout()
plt.show()

# ÄÆ°á»ng dáº«n thÆ° má»¥c chá»©a káº¿t quáº£ detect
# ThÆ° má»¥c nÃ y Ä‘Æ°á»£c táº¡o máº·c Ä‘á»‹nh lÃ  runs/detect/predict sau khi cháº¡y model(..., save=True)
# Náº¿u báº¡n cháº¡y nhiá»u láº§n, nÃ³ sáº½ táº¡o predict2, predict3,...
# Cáº§n tÃ¬m thÆ° má»¥c predict má»›i nháº¥t
detect_runs_dir = "/content/runs/detect"
detect_run_folders = sorted([f for f in os.listdir(detect_runs_dir) if os.path.isdir(os.path.join(detect_runs_dir, f))])
latest_detect_folder = os.path.join(detect_runs_dir, detect_run_folders[-1])
print(f"Latest detect folder: {latest_detect_folder}")


# Hiá»ƒn thá»‹ 4 áº£nh Ä‘áº§u tiÃªn (hoáº·c má»™t vÃ i áº£nh Ä‘áº¡i diá»‡n) tá»« káº¿t quáº£ detect
image_files = sorted(os.listdir(latest_detect_folder))
# Chá»n má»™t sá»‘ áº£nh Ä‘á»ƒ hiá»ƒn thá»‹, vÃ­ dá»¥ 4 áº£nh Ä‘áº§u tiÃªn cÃ³ Ä‘uÃ´i .jpg hoáº·c .png
display_images = [f for f in image_files if f.lower().endswith(('.jpg', '.jpeg', '.png'))][:4]


plt.figure(figsize=(16, 8))
if not display_images:
    print(f"No images found in {latest_detect_folder} to display.")
else:
    for i, img_name in enumerate(display_images):
        img_path = os.path.join(latest_detect_folder, img_name)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(2, 2, i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(img_name)

    plt.tight_layout()
    plt.show()
```
