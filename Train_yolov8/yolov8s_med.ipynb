```python
from ultralytics import YOLO

# Load model
model = YOLO("yolov8s.pt")  # hoặc yolov8m.pt

# Train
model.train(
    data="/content/drive/MyDrive/ML/dataset_balanced/train_data/data.yaml",
    epochs=300,
    imgsz=640,
    batch=32,
    patience=50,
    name="y8s_med",
    project="runs/y8s_med",
    cache=True,
    workers=4,
    optimizer="SGD",  # hoặc thử 'Adam'
    amp=True,
)
```

```python
from ultralytics import YOLO

# Load model đã huấn luyện
model = YOLO("/content/drive/MyDrive/runs/y8s_med/y8s_med2/weights/best.pt")  # hoặc yolov8n.pt, yolov8s.pt..

results = model("/content/drive/MyDrive/ML/dataset_balanced/train_data/images/test", save=True, conf=0.25)

import os
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO

# Đường dẫn thư mục chứa kết quả detect
output_folder = "/content/runs/detect/predict"

# Hiển thị 4 ảnh đầu tiên
image_files = sorted(os.listdir(output_folder))[20:24]

plt.figure(figsize=(16, 8))
for i, img_name in enumerate(image_files):
    img_path = os.path.join(output_folder, img_name)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(2, 2, i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(img_name)

plt.tight_layout()
plt.show()
```

```python
# prompt: hiển thị đồ thị, hình ảnh kết quả đánh giá

# In ra các metric đánh giá
print(metrics)

# Hiển thị đồ thị kết quả huấn luyện
# Các file đồ thị được lưu trong thư mục runs/y8s_med/y8s_medX (X là số lần chạy)
# Ví dụ: F1_curve.png, confusion_matrix.png, results.png, ...
# Tìm thư mục chạy mới nhất
runs_dir = "/content/drive/MyDrive/runs/y8s_med"
run_folders = sorted([f for f in os.listdir(runs_dir) if os.path.isdir(os.path.join(runs_dir, f))])
latest_run_folder = os.path.join(runs_dir, run_folders[-1])
print(f"Latest run folder: {latest_run_folder}")

# Danh sách các file đồ thị bạn muốn hiển thị
plot_files = ["results.png", "F1_curve.png", "confusion_matrix.png", "P_curve.png", "R_curve.png"]

plt.figure(figsize=(18, 12))
for i, plot_file in enumerate(plot_files):
    plot_path = os.path.join(latest_run_folder, plot_file)
    if os.path.exists(plot_path):
        img = cv2.imread(plot_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(2, 3, i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(plot_file)
    else:
        print(f"Plot file not found: {plot_path}")

plt.tight_layout()
plt.show()

# Đường dẫn thư mục chứa kết quả detect
# Thư mục này được tạo mặc định là runs/detect/predict sau khi chạy model(..., save=True)
# Nếu bạn chạy nhiều lần, nó sẽ tạo predict2, predict3,...
# Cần tìm thư mục predict mới nhất
detect_runs_dir = "/content/runs/detect"
detect_run_folders = sorted([f for f in os.listdir(detect_runs_dir) if os.path.isdir(os.path.join(detect_runs_dir, f))])
latest_detect_folder = os.path.join(detect_runs_dir, detect_run_folders[-1])
print(f"Latest detect folder: {latest_detect_folder}")


# Hiển thị 4 ảnh đầu tiên (hoặc một vài ảnh đại diện) từ kết quả detect
image_files = sorted(os.listdir(latest_detect_folder))
# Chọn một số ảnh để hiển thị, ví dụ 4 ảnh đầu tiên có đuôi .jpg hoặc .png
display_images = [f for f in image_files if f.lower().endswith(('.jpg', '.jpeg', '.png'))][:4]


plt.figure(figsize=(16, 8))
if not display_images:
    print(f"No images found in {latest_detect_folder} to display.")
else:
    for i, img_name in enumerate(display_images):
        img_path = os.path.join(latest_detect_folder, img_name)
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(2, 2, i+1)
        plt.imshow(img)
        plt.axis("off")
        plt.title(img_name)

    plt.tight_layout()
    plt.show()
```
