```python
from ultralytics import YOLO

# Load model
model = YOLO("yolov8m.pt")  # hoặc yolov8m.pt

# Train
model.train(
    data="/content/drive/MyDrive/ML/dataset_balanced/train_data/data.yaml",
    epochs=200,
    imgsz=640,
    batch=16,
    patience=60,
    name="y8m_high",
    project="runs/y8m_high",
    cache=True,
    workers=4,
    optimizer="SGD",  # hoặc thử 'Adam'
    amp=True,
    val=True,
    lr0=0.01,
    lrf=0.01,
    momentum=0.937,
    weight_decay=0.0005,
    warmup_epochs=3.0,
    warmup_momentum=0.8,
    box=0.05,
    cls=0.5,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    degrees=0.0,
    translate=0.1,
    scale=0.5,
    shear=0.0,
    perspective=0.0,
    flipud=0.0,
    fliplr=0.5,
    mosaic=1.0,
    mixup=0.05,
    copy_paste=0.1
)
```

```python
import os
import matplotlib.pyplot as plt
import cv2
from ultralytics import YOLO

# Đường dẫn thư mục chứa kết quả detect
output_folder = "/content/runs/detect/predict"

# Hiển thị 4 ảnh đầu tiên
image_files = sorted(os.listdir(output_folder))[20:24]

plt.figure(figsize=(16, 8))
for i, img_name in enumerate(image_files):
    img_path = os.path.join(output_folder, img_name)
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.subplot(2, 2, i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(img_name)

plt.tight_layout()
plt.show()
```

```python


# Các file biểu đồ được lưu trong thư mục runs/y8m_high/y8m_high2
# Đường dẫn tới thư mục chứa các biểu đồ
plot_dir = "/content/drive/MyDrive/runs/y8m_high/y8m_high2"

# Danh sách các file biểu đồ cần hiển thị
plot_files = ["results.png", "confusion_matrix.png", "F1_curve.png", "P_curve.png", "R_curve.png"]

plt.figure(figsize=(16, 10))

for i, plot_file in enumerate(plot_files):
    plot_path = os.path.join(plot_dir, plot_file)
    if os.path.exists(plot_path):
        img = cv2.imread(plot_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

        plt.subplot(2, 3, i+1) # Tạo lưới 2 hàng, 3 cột
        plt.imshow(img)
        plt.title(plot_file)
        plt.axis("off")
    else:
        print(f"File {plot_file} not found at {plot_path}")


plt.tight_layout()
plt.show()
```
